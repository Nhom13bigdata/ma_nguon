# **B1 : TIỀN XỬ LÝ**

```python
# Loại bỏ các cột không cần thiết
df = df.drop(columns = ['Id', 'EmployeeName', 'Notes', 'Status','Agency'])
```

```python
# Xử lý cột số chứa phi số
# hàm chuyển cột số chứa phi số về đúng số
def To_covert_numeric(df, columns):
  df[columns] = pd.to_numeric(df[columns] , errors='coerce')
  return df
# các cột cần chuyển đổi
Columns_convert = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
# chuyển đổi
for column in Columns_convert :
  df = To_covert_numeric(df, column)
# Kiểm tra thông tin sau khi convert
df.info()
```

```python
# tính tỉ lệ khuyết
missing_value = (df.isnull().sum() / len(df))*100
# Convert về dạng DataFrame
missing_value_df = pd.DataFrame(missing_value, columns = ['tỉ lệ khuyết'])
#hiển thị tỉ lệ khuyết
missing_value_df
```

```python
# Xử lý giá trị khuyết với cột BasePay, Benefits
columns_Na = ['BasePay','OvertimePay', 'OtherPay', 'Benefits']
median = df[columns_Na].median()
df[columns_Na] = df[columns_Na].fillna(median)
# Kiểm tra lại giá trị khuyết
df_missing = pd.DataFrame(df.isnull().sum(),columns=['Giá trị khuyết'])
df_missing
```

```python
# xử lý data trùng lặp
row_duplicated = df.duplicated().sum()
print('\nSố lượng data bị trùng : ',row_duplicated)
# Xóa Data trùng lặp
df = df.drop_duplicates()
```

```python
# Xử lý giá trị ngoại lai
def xu_ly_gia_tri_ngoai_lai(df, column):

  # Tính toán IQR (Interquartile Range)
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1

  # Xác định giới hạn trên và dưới để loại bỏ giá trị ngoại lai
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Thay thế các giá trị ngoại lai bằng giới hạn trên và dưới
  df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
  return df[column]

# Xử lý giá trị ngoại lai cho các cột liên quan đến lương
columns_to_process = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'Year']

for column in columns_to_process:
  df[column] = xu_ly_gia_tri_ngoai_lai(df, column)
```

```python
# Kiểm tra số nhãn trong dữ liệu dạng chữ
print('\nKiểm tra số nhãn :')
print(df['JobTitle'], ' chứa ', len(df['JobTitle'].unique()), ' nhãn')

#Lấy 4 nhãn đầu tiên từ cột JobTitle
top_4_labels = df['JobTitle'].unique()[:4]

# Tạo dictionary để map các nhãn
label_mapping = {label: label for label in top_4_labels}
for label in df['JobTitle'].unique():
    if label not in top_4_labels:
        label_mapping[label] = 'Other'

# Áp dụng mapping vào cột 'JobTitle'
df['JobTitle'] = df['JobTitle'].map(label_mapping)

# Mã hóa one-hot encoding
df_encode = pd.get_dummies(df['JobTitle']).astype(int)

# Kết hợp với dữ liệu gốc và loại bỏ cột JobTitle
df_new = pd.concat([df.drop(['JobTitle'], axis=1), df_encode], axis=1)

# Đổi tên cột (chỉ giữ lại 4 nhãn đầu và 'Other')
df_new = df_new.rename(columns={
    'GENERAL MANAGER-METROPOLITAN TRANSIT AUTHORITY': 'Transit Mgr',
    'CAPTAIN III (POLICE DEPARTMENT)': 'Police Captain',
    'WIRE ROPE CABLE MAINTENANCE MECHANIC': 'Cable Mechanic',
    'DEPUTY CHIEF OF DEPARTMENT,(FIRE DEPARTMENT)': 'Deputy Fire Chief'
})

# Sắp xếp lại cột để "Other" ở cuối
cols = [col for col in df_new.columns if col != 'Other'] + ['Other']
df_new = df_new[cols]
```
